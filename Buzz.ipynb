### **1. Data Preprocessing**
- **Handle Missing Values**: Fill empty `share_count`, `reaction_count`, and `comment_count` fields with `0`.
- **Encode Categorical Features**:
  - **One-Hot Encoding**: Apply to `Category`, `Page`, and `Post Type`.
  - **Label Encoding**: Convert `Rating` (e.g., "mostly true", "no factual content") into ordinal labels (e.g., `0`, `1`).
- **Normalize Numerical Features**: Scale `share_count`, `reaction_count`, and `comment_count` to a standard range (e.g., using min-max normalization).

---

### **2. Graph Construction**
- **Nodes**: Represent each post as a node. Each node has features:
  - Encoded `Category`, `Page`, `Post Type`, and `Rating`.
  - Normalized interaction counts (`share_count`, `reaction_count`, `comment_count`).
- **Edges**: Define edges based on **shared attributes**:
  - Connect posts from the **same Page**.
  - Connect posts in the same **Category** (e.g., "mainstream" or "left/right").
  - Optionally, use **temporal edges** to link posts chronologically within the same Page.
- **Edge Weights**: Use interaction metrics (e.g., `share_count`) to weight edges, indicating influence strength.

---

### **3. Model Architecture: GNN + Attention**
Use a **Graph Attention Network (GAT)** with the following components:
- **Graph Attention Layers**:
  - Each node aggregates features from neighbors using attention scores.
  - Attention mechanism calculates weights for neighboring nodes based on feature similarity.
- **Multi-Head Attention**: Capture diverse relationships by using multiple attention heads.
- **Output Layer**: A classification layer to predict `Rating` (e.g., "mostly true" vs. "no factual content").

**Example Architecture**:
```python
GAT(
  in_features = num_node_features,
  hidden_features = 64,
  out_features = num_classes,
  heads = 4,  # Number of attention heads
  dropout = 0.6
)
```

---

### **4. Training and Evaluation**
- **Loss Function**: Cross-entropy loss for classification.
- **Optimizer**: Adam with learning rate `0.001`.
- **Evaluation Metrics**:
  - Accuracy
  - F1-score (to handle class imbalance)
  - Confusion Matrix
- **Train/Test Split**: Split data chronologically (e.g., use older posts for training, newer for testing).

---

### **5. Implementation Tools**
- **Graph Libraries**: Use `PyTorch Geometric` or `Deep Graph Library (DGL)`.
- **Code Skeleton**:
  ```python
  import torch
  from torch_geometric.nn import GATConv

  class GATModel(torch.nn.Module):
      def __init__(self, num_features, num_classes):
          super().__init__()
          self.conv1 = GATConv(num_features, 64, heads=4)
          self.conv2 = GATConv(64*4, num_classes, heads=1)

      def forward(self, data):
          x, edge_index = data.x, data.edge_index
          x = self.conv1(x, edge_index)
          x = torch.relu(x)
          x = self.conv2(x, edge_index)
          return torch.log_softmax(x, dim=1)
  ```

- **Class Imbalance**: Use weighted loss or oversampling for rare labels (e.g., "mixture of true and false").

````python
import pandas as pd
import requests
from tqdm import tqdm  # For progress bar

def extract_context_post(url):
    try:
        # First request to get the redirect URL
        r = requests.get(url, timeout=10)
        r.raise_for_status()

        link_header = r.headers.get('Link', '')
        if not link_header:
            return 0

        redirect_url = link_header[link_header.find('<')+1 : link_header.find('>')]
        if not redirect_url.startswith('http'):
            return 0

        # Second request to the actual content URL
        r2 = requests.get(redirect_url, timeout=10)
        r2.raise_for_status()

        content = r2.content.decode('utf-8', errors='replace')
        start_idx = content.find('meta name="description" content="')

        if start_idx == -1:
            return 0

        start_idx += len('meta name="description" content="')
        end_idx = content.find('" />', start_idx)

        if end_idx == -1:
            return 0

        return content[start_idx:end_idx]

    except Exception as e:
        return 0

# Read CSV file
df = pd.read_csv('facebook-fact-check.csv')

# Add new column with default value 0
df['Context Post'] = 0

# Process each URL with progress bar
for idx, row in tqdm(df.iterrows(), total=len(df)):
    url = row['Post URL']
    if pd.notna(url) and isinstance(url, str) and url.startswith('http'):
        df.at[idx, 'Context Post'] = extract_context_post(url)

# Save updated dataframe
df.to_csv('updated-facebook-fact-check.csv', index=False)
print("Processing completed. Saved to updated-facebook-fact-check.csv")
````
