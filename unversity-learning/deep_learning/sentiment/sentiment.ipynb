```python
import pandas as pd
import re
import os
from nltk.corpus import stopwords
from nltk.stem.snowball import SnowballStemmer
from emoji import demojize
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure you have the necessary NLTK resources
import nltk
nltk.download('stopwords')

# Define the path where your CSV files are located
path = 'extracted_data'  # Replace with your actual path

# List of CSV files
files = ['anger.csv', 'disgust.csv', 'fear.csv', 'joy.csv', 'sad.csv', 'surprise.csv']

# Initialize an empty DataFrame
df = pd.DataFrame()

# Load and concatenate all CSV files, keeping only the 'tweet' and 'emotion' columns
for file in files:
    file_path = os.path.join(path, file)
    temp_df = pd.read_csv(file_path, usecols=['tweet', 'emotion'])
    df = pd.concat([df, temp_df], ignore_index=True)

# Custom Persian stopwords list (example, you may need to refine this list)
persian_stopwords = set([
    'و', 'در', 'به', 'از', 'که', 'این', 'را', 'با', 'برای', 'است', 'آن', 'هم',
    'یک', 'تا', 'می', 'بر', 'بود', 'شد', 'نیز', 'وی', 'ها', 'ما', 'خود', 'من'
])

# Function to remove HTML tags and URLs
def remove_html_urls(text):
    print(f"Before removing HTML and URLs: {text}")
    text = re.sub(r'<.*?>', '', text)  # Remove HTML tags
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)  # Remove URLs
    print(f"After removing HTML and URLs: {text}")
    return text

# Function to remove repetitive characters
def remove_repetitive_characters(text):
    print(f"Before removing repetitive characters: {text}")
    text = re.sub(r'(.)\1+', r'\1', text)  # Replace multiple occurrences of a character with one
    print(f"After removing repetitive characters: {text}")
    return text

# Function to remove stopwords
def remove_stopwords(text):
    print(f"Before removing stopwords: {text}")
    words = text.split()
    words = [word for word in words if word not in persian_stopwords]
    result = ' '.join(words)
    print(f"After removing stopwords: {result}")
    return result

# Function to replace emojis with their textual equivalents
def replace_emojis(text):
    print(f"Before replacing emojis: {text}")
    result = demojize(text)
    print(f"After replacing emojis: {result}")
    return result

# Initialize stemmer
stemmer = SnowballStemmer("english")

# Function to apply stemming
def apply_stemming(text):
    print(f"Before stemming: {text}")
    words = text.split()
    words = [stemmer.stem(word) for word in words]
    result = ' '.join(words)
    print(f"After stemming: {result}")
    return result

# Function to correct spelling (stub function, needs implementation)
def correct_spelling(text):
    # Implement a spelling correction function or use a library/tool for this
    print(f"Before correcting spelling: {text}")
    # Assuming a placeholder function that returns the text as-is
    result = text
    print(f"After correcting spelling: {result}")
    return result

# Sample tweets for demonstration
sample_tweets = [
    "این یک <b>تست</b> است http://example.com",  # Example for HTML and URLs
    "خووووووب",  # Example for repetitive characters
    "من به مدرسه می‌روم",  # Example for stopwords
    "من 😊 خوشحالم",  # Example for emojis
    "کتاب‌ها خوانده شده‌اند"  # Example for stemming
]

# Apply preprocessing steps to sample tweets
for tweet in sample_tweets:
    tweet = remove_html_urls(tweet)
    tweet = remove_repetitive_characters(tweet)
    tweet = replace_emojis(tweet)
    tweet = remove_stopwords(tweet)
    tweet = apply_stemming(tweet)
    tweet = correct_spelling(tweet)
    print("\n")

# Apply preprocessing steps to the actual dataset
df['tweet'] = df['tweet'].apply(remove_html_urls)
df['tweet'] = df['tweet'].apply(remove_repetitive_characters)
df['tweet'] = df['tweet'].apply(replace_emojis)
df['tweet'] = df['tweet'].apply(remove_stopwords)
df['tweet'] = df['tweet'].apply(apply_stemming)
df['tweet'] = df['tweet'].apply(correct_spelling)

# Display the first few rows of the DataFrame to verify the preprocessing
print(df.head())

# Plot the distribution of classes
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='emotion', palette='viridis')
plt.title('Distribution of Emotions in the Dataset')
plt.xlabel('Emotion')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
