```python
import numpy as np
import matplotlib.pyplot as plt

# Function to load dataset from .dat file
def load_dataset(file_path):
    with open(file_path, 'r') as file:
        data = [list(map(float, line.strip().split())) for line in file]
    return np.array(data)

# Function to pad or truncate rows to ensure uniformity
def preprocess_data(dataset):
    max_len = max(len(row) for row in dataset)
    processed_data = np.array([row + [0] * (max_len - len(row)) for row in dataset])
    return processed_data

# Function to split dataset into input and output
def split_dataset(dataset):
    X = dataset[:, :-1]
    y = dataset[:, -1]
    return X, y


# Radial Basis Function (RBF) implementation
def rbf(x, centers, sigma):
    x_reshaped = np.atleast_2d(x)
    centers_reshaped = np.atleast_2d(centers)

    return np.exp(-np.sum((x_reshaped - centers_reshaped)**2, axis=1) / (2 * sigma**2))

# Function to train RBF network
def train_rbf(X, y, num_centers, sigma, learning_rate, epochs):
    # Initialize weights and centers randomly
    weights = np.random.rand(num_centers)
    centers = np.random.rand(num_centers, X.shape[1])

    for epoch in range(epochs):
        for i in range(X.shape[0]):
            phi = rbf(X[i], centers, sigma)
            prediction = np.dot(phi, weights)
            error = y[i] - prediction
            weights += learning_rate * error * phi

    return centers, weights

# Function to predict output using RBF network
def predict_rbf(X, centers, weights, sigma):
    phi = np.array([rbf(x, centers, sigma) for x in X])
    predictions = np.dot(phi, weights)
    return np.sign(predictions)

# Function to calculate accuracy
def calculate_accuracy(predictions, targets):
    return np.mean(predictions == targets)

# Function to plot accuracy
def plot_accuracy(train_acc, test_acc):
    epochs = len(train_acc)
    plt.plot(range(1, epochs + 1), train_acc, label='Training Accuracy')
    plt.plot(range(1, epochs + 1), test_acc, label='Test Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()

# Load train and test datasets
train_data = load_dataset('spiral_100.dat')
test_data = load_dataset('spiral_g.dat')

# Preprocess data to handle varying number of elements
train_data = preprocess_data(train_data)
test_data = preprocess_data(test_data)

# Split datasets into input and output
X_train, y_train = split_dataset(train_data)
X_test, y_test = split_dataset(test_data)

# Set hyperparameters
num_centers = 10  # Number of RBF centers
sigma = 1.0      # RBF width
learning_rate = 0.01
epochs = 50

# Train RBF network
centers, weights = train_rbf(X_train, y_train, num_centers, sigma, learning_rate, epochs)

# Predict on training and test datasets
train_predictions = predict_rbf(X_train, centers, weights, sigma)
test_predictions = predict_rbf(X_test, centers, weights, sigma)

# Calculate accuracy
train_accuracy = calculate_accuracy(train_predictions, y_train)
test_accuracy = calculate_accuracy(test_predictions, y_test)

# Print output and accuracy
print("Training Data:")
print("Input:", X_train)
print("Output:", y_train)
print("Predictions:", train_predictions)
print("Training Accuracy:", train_accuracy)

print("\nTest Data:")
print("Input:", X_test)
print("Output:", y_test)
print("Predictions:", test_predictions)
print("Test Accuracy:", test_accuracy)

# Plot accuracy
plot_accuracy(train_accuracy, test_accuracy)
```
