```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x, a, c):
    return 1 / (1 + np.exp(-a * (x - c)))

x = np.linspace(-5, 5, 100)  # Generate a range of x values
a = 1  # Sigmoid slope parameter(shib)
c = 0  # Sigmoid center parameter(markaz)

y = sigmoid(x, a, c)  # Compute the sigmoid values

plt.plot(x, y)  # Plot the sigmoid curve
plt.xlabel('x')
plt.ylabel('Membership')
plt.title('Sigmoid Membership Function')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def membership_function(x, left, right):
    """
    Generates a left-right membership function.
    Parameters:
        - x: Input values.
        - left: Left boundary value.
        - right: Right boundary value.
    Returns:
        - y: Membership values corresponding to the input values.
    """
    y = np.zeros_like(x)
    y[x < left] = 1
    y[(x >= left) & (x <= right)] = (right - x[(x >= left) & (x <= right)]) / (right - left)
    return y

# Generate x values
x = np.linspace(0, 10, 100)

# Define left and right boundaries
left = 2
right = 7

# Generate membership function values
y = membership_function(x, left, right)

# Plot the membership function
plt.plot(x, y)
plt.xlabel('Input')
plt.ylabel('Membership')
plt.title('Left-Right Membership Function')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the membership functions
def FL_membership_function(x):
    return np.sqrt(np.maximum(0, 1 - x**2))

def FR_membership_function(x):
    return np.exp(-x)**3

# Generate x values
x = np.linspace(-1, 1, 100)

# Calculate membership values
FL_values = FL_membership_function(x)
FR_values = FR_membership_function(x)

# Plot the FL membership function
plt.subplot(2, 1, 1)
plt.plot(x, FL_values)
plt.xlabel('x')
plt.ylabel('Membership')
plt.title('FL Membership Function')

# Plot the FR membership function
plt.subplot(2, 1, 2)
plt.plot(x, FR_values)
plt.xlabel('x')
plt.ylabel('Membership')
plt.title('FR Membership Function')

plt.tight_layout()
plt.show()
```
