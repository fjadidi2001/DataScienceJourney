## Handle Missing Values

```python
# Assuming df is your DataFrame
numeric_columns = df.select_dtypes(include=np.number).columns
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())  # Fill numerical NaNs with mean

categorical_columns = df.select_dtypes(exclude=np.number).columns
df[categorical_columns] = df[categorical_columns].fillna(df[categorical_columns].mode().iloc[0])  # Fill categorical NaNs with mode
```

## Encode Categorical Variables

```python
from sklearn.preprocessing import LabelEncoder

# Apply Label Encoding to categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, GRU, Conv1D, Flatten, MaxPooling1D

def build_rnn_model(input_shape):
    model = Sequential([
        LSTM(64, return_sequences=True, input_shape=input_shape),
        LSTM(32),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mse')
    return model

def build_gru_model(input_shape):
    model = Sequential([
        GRU(64, return_sequences=True, input_shape=input_shape),
        GRU(32),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mse')
    return model

def build_cnn_model(input_shape):
    model = Sequential([
        Conv1D(64, 2, activation='relu', input_shape=input_shape),
        MaxPooling1D(pool_size=2),
        Flatten(),
        Dense(32, activation='relu'),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mse')
    return model

# Reshape input data for RNN/CNN models
X_train_rnn = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_val_rnn = X_val.reshape((X_val.shape[0], X_val.shape[1], 1))

# Train RNN Model
rnn_model = build_rnn_model((X_train_rnn.shape[1], X_train_rnn.shape[2]))
rnn_model.fit(X_train_rnn, y_train, validation_data=(X_val_rnn, y_val), epochs=10, batch_size=32, verbose=0)
rnn_loss = rnn_model.evaluate(X_val_rnn, y_val, verbose=0)

# Train GRU Model
gru_model = build_gru_model((X_train_rnn.shape[1], X_train_rnn.shape[2]))
gru_model.fit(X_train_rnn, y_train, validation_data=(X_val_rnn, y_val), epochs=10, batch_size=32, verbose=0)
gru_loss = gru_model.evaluate(X_val_rnn, y_val, verbose=0)

# Train CNN Model
cnn_model = build_cnn_model((X_train_rnn.shape[1], X_train_rnn.shape[2]))
cnn_model.fit(X_train_rnn, y_train, validation_data=(X_val_rnn, y_val), epochs=10, batch_size=32, verbose=0)
cnn_loss = cnn_model.evaluate(X_val_rnn, y_val, verbose=0)
```

```python
# 1. Linear Regression
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
y_pred_train_lin = lin_reg.predict(X_train)
y_pred_val_lin = lin_reg.predict(X_val)

# 2. XGBoost
xgb_model = XGBRegressor()
xgb_model.fit(X_train, y_train, eval_set=[(X_val, y_val)], verbose=False)
y_pred_train_xgb = xgb_model.predict(X_train)
y_pred_val_xgb = xgb_model.predict(X_val)

# 3. TabNet
tabnet_model = TabNetRegressor()
tabnet_model.fit(X_train, y_train.values.reshape(-1, 1), # Reshape y_train to 2D
                 eval_set=[(X_val, y_val.values.reshape(-1, 1))], # Reshape y_val as well
                 max_epochs=50, patience=10, batch_size=1024, virtual_batch_size=128)
y_pred_train_tabnet = tabnet_model.predict(X_train).squeeze()
y_pred_val_tabnet = tabnet_model.predict(X_val).squeeze()

# 4. Random Forest
rf_model = RandomForestRegressor()
rf_model.fit(X_train, y_train)
y_pred_train_rf = rf_model.predict(X_train)
y_pred_val_rf = rf_model.predict(X_val)

# 5. RNN
model_rnn = Sequential([
    tf.keras.layers.Reshape((X_train.shape[1], 1), input_shape=(X_train.shape[1],)),
    LSTM(64, return_sequences=True),
    LSTM(32),
    Dense(1)
])
model_rnn.compile(optimizer='adam', loss='mse', metrics=['mae'])
history_rnn = model_rnn.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=100, batch_size=64)

# 6. CNN
model_cnn = Sequential([
    tf.keras.layers.Reshape((X_train.shape[1], 1), input_shape=(X_train.shape[1],)),
    Conv1D(64, kernel_size=2, activation='relu'),
    Flatten(),
    Dense(32, activation='relu'),
    Dense(1)
])
model_cnn.compile(optimizer='adam', loss='mse', metrics=['mae'])
history_cnn = model_cnn.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=100, batch_size=64)

# 7. GRU
model_gru = Sequential([
    tf.keras.layers.Reshape((X_train.shape[1], 1), input_shape=(X_train.shape[1],)),
    GRU(64, return_sequences=True),
    GRU(32),
    Dense(1)
])
model_gru.compile(optimizer='adam', loss='mse', metrics=['mae'])
history_gru = model_gru.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=100, batch_size=64)
```

```python
# Plotting loss and accuracy for deep learning models
def plot_loss(history, title):
    plt.figure(figsize=(10, 6))
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f'{title} Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()

plot_loss(history_rnn, "RNN")
plot_loss(history_cnn, "CNN")
plot_loss(history_gru, "GRU")

# Plotting performance for machine learning models
def plot_model_performance(model_name, y_train_pred, y_val_pred):
    train_mse = mean_squared_error(y_train, y_train_pred)
    val_mse = mean_squared_error(y_val, y_val_pred)

    print(f"{model_name} Train MSE: {train_mse}")
    print(f"{model_name} Validation MSE: {val_mse}")

    sns.scatterplot(x=y_train, y=y_train_pred, label='Train')
    sns.scatterplot(x=y_val, y=y_val_pred, label='Validation')
    plt.title(f'{model_name} Predictions')
    plt.xlabel('True Values')
    plt.ylabel('Predicted Values')
    plt.legend()
    plt.show()

plot_model_performance("Linear Regression", y_pred_train_lin, y_pred_val_lin)
plot_model_performance("XGBoost", y_pred_train_xgb, y_pred_val_xgb)
plot_model_performance("TabNet", y_pred_train_tabnet, y_pred_val_tabnet)
plot_model_performance("Random Forest", y_pred_train_rf, y_pred_val_rf)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

def evaluate_model_performance(model_name, y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)

    print(f"{model_name} Performance on Test Set:")
    print(f"  Mean Squared Error (MSE): {mse:.4f}")
    print(f"  Root Mean Squared Error (RMSE): {rmse:.4f}")
    print(f"  R-squared (R²): {r2:.4f}")
    print("-" * 50)

# 1. Linear Regression
y_test_pred_lin = lin_reg.predict(X_test)
evaluate_model_performance("Linear Regression", y_test, y_test_pred_lin)

# 2. XGBoost
y_test_pred_xgb = xgb_model.predict(X_test)
evaluate_model_performance("XGBoost", y_test, y_test_pred_xgb)

# 3. TabNet
y_test_pred_tabnet = tabnet_model.predict(X_test).squeeze()
evaluate_model_performance("TabNet", y_test, y_test_pred_tabnet)

# 4. Random Forest
y_test_pred_rf = rf_model.predict(X_test)
evaluate_model_performance("Random Forest", y_test, y_test_pred_rf)

# 5. RNN
y_test_pred_rnn = model_rnn.predict(X_test).squeeze()
evaluate_model_performance("RNN", y_test, y_test_pred_rnn)

# 6. CNN
y_test_pred_cnn = model_cnn.predict(X_test).squeeze()
evaluate_model_performance("CNN", y_test, y_test_pred_cnn)

# 7. GRU
y_test_pred_gru = model_gru.predict(X_test).squeeze()
evaluate_model_performance("GRU", y_test, y_test_pred_gru)
```
